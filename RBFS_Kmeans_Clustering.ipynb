#  K = 12

```python
def distance(x0,y0,x1,y1):
    return np.sqrt((x1-x0)**2 + (y1-y0)**2)
```

```python
n=12
c=[]
while 1:
    a=np.random.choice(range(len(X)), n)
    #print(a)
    ln=len(set(a))
    if ln==n:
        for i in range(n):
            c.append(X[a[i]])
        break
c=np.array(c)
#print(c)
```

```python
n=len(X)
dis={}
for i in range(len(c)):
    dis[i]=[]

for i in range(n):
    mn=-1
    mx=99999999
    for j in range(len(c)):
        d_id=distance(X[i][0],X[i][1],c[j][0],c[j][1])
        if d_id < mx:
            mx=d_id
            mn=j
    dis[mn].append(X[i])

for i in range(len(c)):
    avg_x,avg_y=0,0
    temp=dis[i]
    
    for j in range(len(temp)):
        avg_x+=temp[j][0]
        avg_y+=temp[j][1]
        
    avg_x/=len(temp)
    avg_y/=len(temp)
    c[i]=[avg_x,avg_y]

c=np.array(c)
```

```python
print(c)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
colmap = {0: 'r', 1: 'g', 2: 'y', 3:'b'}
for i in range(12):
    plt.scatter(c[i][0], c[i][1], color=colmap[0])
plt.show()
```
